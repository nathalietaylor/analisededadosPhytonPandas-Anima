```python
# Importando biblioteca pandas no Python
import pandas as pd
```

```python
combustiveis_df = pd.read_excel('ca-2021-02.xlsx')
display(combustiveis_df)
```

```python
#Inserção simples de dados
combustiveis_df['Ativo'] = True
display(combustiveis_df.head())
```

```python
# Criar uma coluna "Obs" que tenha nela escrito "MELHOR CIDADE" quando a coluna Municipio for SAO PAULO
combustiveis_df['Obs'] = ['MELHOR CIDADE' if Municipio == 'SAO PAULO' else'' for Municipio in combustiveis_df['Municipio']]
display(combustiveis_df.loc[combustiveis_df['Municipio'].isin(['SAO PAULO', 'INDAIATUBA', 'CAMPINAS']),['Municipio','Obs']])
```

```python
# como preencher uma coluna "Valor de Venda - Status" que verifica o seguinte:
#se o valor de venda for maior que 6,5 reais, ele fala que está caro, caso contrario, está barato
import numpy as np

combustiveis_df['Valor de Venda - Status'] = np.where(combustiveis_df['Valor de Venda'] > 6.5, "Caro", "Barato")
display(combustiveis_df[['Revenda', 'Valor de Venda', 'Valor de Venda - Status']])
```

```python
# Calcular postos de gasolina por habitante temos na amostragem de combustiveis nov/2021
num_habitante_df = pd.read_csv('ibge_num_habitantes_estimado.csv', sep=";")
num_habitante_df.rename(columns={'Estado':'Estado - Sigla'}, inplace = True)
display(num_habitante_df)
```

```python
# Faz um merge dos dois dataframes
colunas = ['Municipio', 'Estado - Sigla']
merge_df = combustiveis_df.merge(num_habitante_df, how="inner", on=colunas)
display(merge_df)
print(merge_df.info())
```

```python
#Destruir coluna completamente vazia (todas as linhas são nulas)
merge_df.dropna(axis='columns', inplace=True)
print(merge_df.info())
```

```python
colunas=['Regiao - Sigla', 'Nome da Rua', 'Numero Rua', 
         'Bairro', 'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
         'Unidade de Medida', 'Bandeira', 'Ativo', 'Valor de Venda - Status']
merge_df.drop(labels=colunas, axis=1, inplace=True)
print(merge_df.info())
```

```python
# Remover as linhas duplicadas
merge_df.drop_duplicates(inplace=True)
display(merge_df.head(100))
```

```python
#Agrupar e contar quantos postos tem na cidade
postos_por_municipio_df = merge_df.groupby(by=['Estado - Sigla', 'Municipio', 'NumHabitantes2021']).count()
postos_por_municipio_df.drop('CNPJ da Revenda', axis=1, inplace=True)
postos_por_municipio_df.rename(columns={'Revenda': 'Numero de Postos'}, inplace=True)
postos_por_municipio_df.reset_index(inplace=True)
display(postos_por_municipio_df)
```

```python
#Calcular nº habitantes/postos
postos_por_municipio_df['HabitantesPorPostos'] = postos_por_municipio_df ['NumHabitantes2021']/postos_por_municipio_df['Numero de Postos']
postos_por_municipio_df['PostosPorHabitantes'] = postos_por_municipio_df['Numero de Postos']/postos_por_municipio_df ['NumHabitantes2021']
display(postos_por_municipio_df)
```

```python
#Para saber quantas vendas foram realizas no município de São Paulo (?NT)
word = "SAO PAULO"
count = combustiveis_df[combustiveis_df["Municipio"].str.contains(word)].shape[0]
display(count)
```

```python
#Vamos criar gráficos

import matplotlib.pyplot as plt

plt.hist(combustiveis_df['Valor de Venda'])
#Para colocar título no gráfico
plt.title("Valor médio dos combustíveis Nov/2021")
#Rótulo horizontal e vertical
plt.xlabel('Preço (em reais)')
plt.ylabel('Quantidade de dados levantados')

#Traça a linha vermelha tracejada com o preço médio
plt.axvline(combustiveis_df['Valor de Venda'].mean(), color='red', linestyle='dashed', linewidth=2)

#Plotar gráfico
plt.show()
```

```python
#Segundo gráfico
c_mean = combustiveis_df['Valor de Venda'].groupby(by=combustiveis_df['Produto']).mean()
display(c_mean)
```

```python
#outra biblioteca para customizar os gráficos --> seaborn
import seaborn as sns

#Definir área do gráfico
plt.figure(figsize=(7,5))
#Plotar gráfico
c_mean.plot(
    kind="barh",
    xlabel="Tipo de Combustível",
    ylabel="Preço reais/litro", color='red', 
    title="Média de preços por combustível",
  #para ajustar opacidade da barra
    alpha=0.8
)
#Grid
plt.grid()

#Remover as linhas superior e lateral direita do gráfico 
sns.despine()

#Exibir
plt.show()
```

```python
#Para personalizar a planilha excel
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Color, Alignment 

#para abrir o excel no openpyxl
wb = load_workbook(excel) # wb = workbook

#pegar a planilha certa, usando o sheet name
ws = wb['Sumário'] # work sheet -> planilha atual, ativa, de trbalho

#pintar o cabeçalho da tabela de cinza claro
cinzinha = PatternFill("solid", fgColor="CCCCCC")
coords = ['A1', 'B1']
for coord in coords:
  ws[coord].fill = cinzinha

#pintar a fonte de vermelho e deixar negrito quando preço do combustível for maior ou igual a 6,5 reais
MAX_ROW = ws.max_row
num_linha =2
while (num_linha <= MAX_ROW):
  coord = 'B' +str(num_linha)
  if ws[coord].value >= 6.5:
     ws[coord].font = Font(bold=True, color="FF0000")
  num_linha = num_linha + 1

#Salvar o Excel
wb.save(excel)
```
